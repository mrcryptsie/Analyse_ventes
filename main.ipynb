```python
# Préparation des données pour l'apprentissage automatique
# Néttoyage et prétaitement
# Importation du module de preprocessing
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import MinMaxScaler

# Transformer la variable cible "bank_account " en variable numérique
le = LabelEncoder()
df['bank_account'] = le.fit_transform(df['bank_account'])
df['bank_account']
```

```python
# Séparer les variables indépendantes de la variable target
# Variables indépendantes
x_train = df.drop(['bank_account'], axis=1)
# Variables cible ou target
y_train = df['bank_account']
```

```python
def preprocessing_data(df):
    # Convertir les entiers en floats
    float_array = df[['year', 'household_size', 'age_of_respondent']].values.astype(float)
    # Convertir à présent les variables strings en floats ou numériques
    # Recenssons d'abord ces variables catégoriqes dans une matrice
    var_string = ['country', 'relationship_with_head', 'marital_status', 'education_level', 'job_type']
    # Utiliser One Hot Encoding pour transformer les variables en numériques
    df = pd.get_dummies(df, prefix_sep="_", columns=var_string)
    # Utiliser Label Encoder pour gérer les valeurs numériques
    le = LabelEncoder()
    df['location_type'] = le.fit_transform(df['location_type'])
    df['cellphone_access'] = le.fit_transform(df['cellphone_access'])
    df['gender_of_respondent'] = le.fit_transform(df['gender_of_respondent'])
    # Effacer la colonne uniqueid
    df = df.drop(['uniqueid'], axis=1)
    # Effectuer la mise à l'échelle des fonctionnalités
    scaler = MinMaxScaler(feature_range=(0, 1))
    df = pd.DataFrame(scaler.fit_transform(df), columns=df.columns)
    return df
# Stocker le résultat de la fonction dans une variable
df_transformed = preprocessing_data(df)
# Afficher les premières lignes du dataframe transformé
df_transformed
```

```python
# Phase de préprocessus du modèle d'entrainement et de test
processed_train = preprocessing_data(x_train)
processed_test = preprocessing_data(df_test)
```

```python
# Obseravation de notre première entrainement
print(processed_train[:1])
```

```python
# Voir la dimenssion de notre modèle d'entrainement
processed_train.shape
```

```python
# Split ou fendre les données d'entrainement
# Importer le module nécessaire
from sklearn.model_selection import train_test_split
X_Train, X_Val, y_Train, y_val = train_test_split(processed_train, y_train, stratify=y_train, test_size=0.1, random_state=42)
```

***Il s'agit d'une classification:***
Nous allons pouvoir utiliser cinq algorithmes de classification pour repondre à cette belle préoccupations:


*   Foret aléatoire
*   K voisins les plus proches
*   Regression logistiques
*   Arbres supplémentaires 
*   XGBoost





```python
# Importer mes modèles de classifications à utiliser
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import ExtraTreesClassifier
from xgboost import XGBClassifier
```

```python
# Création des modèles respectifs
lg_model = LogisticRegression()
rf_model = RandomForestClassifier()
kn_model = KNeighborsClassifier()
et_model = ExtraTreesClassifier()
xg_model = XGBClassifier()
```

```python
# Entrainer le modèle avec a méthode fit
lg_model.fit(X_Train, y_Train)
rf_model.fit(X_Train, y_Train)
kn_model.fit(X_Train, y_Train)
et_model.fit(X_Train, y_Train)
xg_model.fit(X_Train, y_Train)
```

```python
# Importer la bibliothèque d'évaluation du metrics
from sklearn.metrics import confusion_matrix, accuracy_score

# Evaluer le model
lg_y_model = lg_model.predict(X_Val)
rf_y_model = rf_model.predict(X_Val)
kn_y_model = kn_model.predict(X_Val)
et_y_model = et_model.predict(X_Val)
xg_y_model = xg_model.predict(X_Val)
```
